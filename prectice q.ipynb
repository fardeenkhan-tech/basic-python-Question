```python
# count vowels and consonants 

s = "fardeenis a good boy "

v = 0
c = 0 

for ch in s :
  if ch.lower() in "aeiou" :
    v += 1 
  else:
    c += 1 

print(v)
print(c)
```

```python
# prime number 
num =  9

cont = 0 

for i in range(1,num+1):

  if num % i == 0 :
    cont += 1 

if cont == 2 :
  print("prime number")

else:
  print("not prime number")
```

```python
# Character Frequency 

s = "my name is fardeen khan and i am computer science engineer by education with strong and logical thinking and i have appliyed my experrice into data science and data analiysis"

freq = {}

for ch in s :
  if ch not in freq :
    freq[ch] = 1 
  else:
    freq[ch] += 1 

print(freq)
```
