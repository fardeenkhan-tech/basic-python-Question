```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


root = Node(10)
root.left = Node(5)
root.right = Node(20)
```

```python
# Tree question 

def count_nodes(root):
  if root is None:
    return 0
  

  return 1 + count_nodes(root.left) + count_nodes(root.right)


```

```python
count_nodes(root)
```

```python
count_nodes(root.right)
```

```python
count_nodes(root.left)
```

```python
count_nodes(root.right) + count_nodes(root.left)
```

```python
def sum_nodes(root):
  if root is None:
    return 0 
  
  return root.data + sum_nodes(root.left) + sum_nodes(root.right)
```

```python
print(sum_nodes(root))
```

```python
#  count even-valued nodes 


def count_even(root):
  if root is None:
    return 0 
  
  count = 1  if root.data % 2 == 0 else 0 
  return count + count_even(root.left) + count_even(root.right)
```

```python
print(count_even(root))
```

```python
# find maximum value in a binary tree 


def find_max(root):
  if root is None :
    return float("-inf")


  return max(
   root.data,
   find_max(root.left),
   find_max(root.right)
)
```

```python
print(find_max(root))
```

```python
# find min value in a binary tree 



def  find_min(root):
  if root is None :
    return float("inf")
  

  return min(
    root.data,
    find_min(root.left),
    find_min(root.right)
  )
```

```python
print(find_min(root))
```

```python
#  serarch element in a binary tree is exists or not 

def search(root,key):
  if root is None :
    return False 
  
  if root.data == key:
    return True 
  

  return search(root.left, key) or search(root.right,key)
```

```python
print(search(root,10))
```

```python
#  find even lagest number 

n = [10,1,3,4,5,30,50,100,800]

largest = None

for num in n :
  if num % 2 == 0 :
    if largest is None or num > largest :
      largest = num

print(largest)
```

```python
# seprate even and odd numbers 

n = [10,1,3,4,5,30,50,100,800]

even = []
odd = []

for num in n :
  if  num %2 ==0 :
    even.append(num)
  else :
    odd.append(num)



print(even)
print(odd)
```

```python
# even numbers using list comprehension 

n = [10,30,4,0,49,89,48,23,22,55,77,100]

even = [num for num in n if num %2==0]
print(even)
```
